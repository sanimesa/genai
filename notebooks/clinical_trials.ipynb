```python
api_endpoint = 'https://clinicaltrials.gov/api/v2/studies'
params = {'query.cond': 'sickle cell', 'query.locn': 'United States',
          'query.patient': 'FEMALE ADULT',
          'filter.overallStatus': 'RECRUITING', 'countTotal': 'true'}
```

```python
# prompt: check the header of the api

response = requests.get(api_endpoint, params=params)
print(response.headers)
print(response.status_code)
```

```python
response.json()['totalCount'], len(response.json()['studies'])
```
